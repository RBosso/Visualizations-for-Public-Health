```python
#COVID Correlations
#Import Libraries Used
#The Datasets used are from:

# Hannah Ritchie, Edouard Mathieu, Lucas Rod√©s-Guirao, Cameron Appel, Charlie Giattino, 
# Esteban Ortiz-Ospina, Joe Hasell, Bobbie Macdonald, Diana Beltekian and Max Roser (2020) 
# - "Coronavirus Pandemic (COVID-19)". Published online at OurWorldInData.org. 
# Retrieved from: 'https://ourworldindata.org/coronavirus' [Online Resource]

import os
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt

plt.style.use('ggplot')

from matplotlib.pyplot import figure

%matplotlib inline 

#Reads from the dataset
df = pd.read_csv(r'/content/sample_data/diabetes_and_Mortality.csv')
df.head() #Top of the dataset
```

```python
plt.scatter(x = df['diabetes_prevalence'], y = df['COVID_Mortality_Rate'])
plt.title('Diabetes Prevalence vs. COVID Mortality Rate')
plt.xlabel('Diabetes Prevalence')
plt.ylabel('COVID Mortality Rate')
plt.show()
df.corr()
```

```python
#Overall, when considering COVID mortality rates across multiple countries,
#there seems to be a poor correlation between mortality rates and diabetes prevalence.
```

```python
#Reads from the dataset
df = pd.read_csv(r'/content/sample_data/GDP_and_Mortality.csv')
df.head() #Top of the dataset
```

```python
plt.scatter(x = df['gdp_per_capita'], y = df['COVID_Mortality_Rate'])
plt.title('GDP vs. COVID Mortality Rate')
plt.xlabel('GDP Per Capita')
plt.ylabel('COVID Mortality Rate')
plt.show()
```

```python
sns.regplot(x = 'gdp_per_capita', y='COVID_Mortality_Rate', data = df, scatter_kws = {"color" : "red"}, line_kws = {"color" : "blue"})
df.corr()
```

```python
matrix = df.corr(method = "spearman") #Monotonic
sns.heatmap(matrix, annot = True)
plt.title("Correlation Matrix for Features Using Spearman Method")
plt.show()
print()
matrix = df.corr(method = "pearson") #Linear
sns.heatmap(matrix, annot = True)
plt.title("Correlation Matrix for Features Using Pearson Method")
plt.show()
```

```python
#There does seem to be some correlation between GDP Per Capita and COVID Mortality Rates.
```

```python
#Reads from the dataset
df = pd.read_csv(r'/content/sample_data/human_development_index_and_GDP.csv')
df.head() #Top of the dataset
```

```python
plt.scatter(x = df['human_development_index'], y = df['gdp_per_capita'])
plt.title('Human Development Index vs. GDP')
plt.xlabel('Human Development')
plt.ylabel('GDP Per Capita')
plt.show()
```

```python
sns.regplot(x = 'human_development_index', y='gdp_per_capita', data = df, scatter_kws = {"color" : "red"}, line_kws = {"color" : "blue"}, order = 2)
df.corr()
#A second - order polynomial regression fits this data very well
```

```python
matrix = df.corr(method = "spearman") #Monotonic
sns.heatmap(matrix, annot = True)
plt.title("Correlation Matrix for Features Using Spearman Method")
plt.show()
print()
matrix = df.corr(method = "pearson") #Linear
sns.heatmap(matrix, annot = True)
plt.title("Correlation Matrix for Features Using Pearson Method")
plt.show()
```

```python
# In addition to the fact that there is a strong linear correlation between GDP and Human Development Index,
# there is an even stronger correlation indicated by the Spearman Method, suggesting that the relation
# between GDP Per Capita and the Human Development Index is more effectively modeled by a nonlinear function. 
```
